**Operations_sheet**

```python
opr_sheet = pd.read_excel('Data Science Case study Vindiata.xls',sheet_name = 'formatted_Operations')
col_names = ['Jan-14','Feb-14','Mar-14','Apr-14','May-14','Jun-14','Jul-14','Aug-14','Sep-14','Oct-14','Nov-14','Dec-14']
opr_sheet.rename(columns=dict(zip(opr_sheet.columns[2:], col_names)), inplace=True)
opr_sheet
```

**City_pairs_sheet**

```python
cp_sheet = pd.read_excel('Data Science Case study Vindiata.xls',sheet_name = 'formatted_city_pairs')
cp_sheet.rename(columns = {'Pass. Demand \n(per day)':'Pass. Demand (per day)'}, inplace = True)
cp_sheet
```

```python
#creating a new column "Total_Hr_Flown" in "opr_sheet":

opr_sheet['Total_Hr_Flown'] = opr_sheet.iloc[:,2:].sum(axis = 1)
opr_sheet.head()
```

```python
#creating a copy of dataframe "opr_sheet" so that the changes won't reflect to the original dataframe

opr_sheet_new = opr_sheet.copy()
opr_sheet_new = opr_sheet_new.groupby(['Aircraft Type']).sum()

#merging 2 dataframes i.e. "Ac_characterstics" and "Operations"

new_df = df.merge(opr_sheet_new, how = 'left', on = 'Aircraft Type')
new_df
```

### 1) Total Cost As Per Aircraft Type

#### Total cost = Hours flown * Costs / hour


```python
new_df['Total_Cost(USD)'] = new_df['Total_Hr_Flown'] * new_df['Costs per flight hour']
new_df[['Aircraft Type','Total_Cost(USD)' ]]
```

### 2) Cost per seat per Km by Aircraft Type :


**Cost per seat per km = Total cost(USD) / (No. of seats x Distance Covered in (km) )  ------------------------ 1)**

**Distance Covered in (km) : Total hours x Avg. speed   ----------------------------------2)**



* Substituting equation 2) in equation 1):

* New Formula Derieved: 

**Cost per seat per km = Total cost(USD) / (No. of seats x Total hours x Avg. speed)**





```python
new_df['Cost per seat per km'] = new_df['Total_Cost(USD)'] / (new_df['Number of Seats'] * new_df['Total_Hr_Flown'] * new_df['Ave. Speed (km/h)'])

## sorting the "Cost per seat per km" column in ascending order :
new_df[['Aircraft Type', 'Cost per seat per km']].sort_values(by = ['Cost per seat per km'])
```

**Conclusion** : Aircraft  **A330** is the best suited aircraft type as it the most cost effective aircraft which costs **0.033333 USD** only.

**City Pair AA-BB**

```python
df1 = new_df.iloc[:,:4].copy()

# Travel cost and remaining passengers after 1st flight
df1['AA-BB cost 1'] = np.where(df1['Range (Km)'] >= 3000, (df1['Cost per seat per km'] * df1['Number of Seats'] * 3000).astype(int), 0)
df1['AA-BB remain 1'] = np.where(np.logical_and(420 - df1['Number of Seats'] >= 0, df1['AA-BB cost 1'] != 0), 420 - df1['Number of Seats'], 0)

# Total travel cost and remaining passengers after 2nd flight
df1['AA-BB cost 2'] = np.where(df1['Range (Km)'] >= 3000, (df1['Cost per seat per km'] * df1['Number of Seats'] * 3000).astype(int) + df1['AA-BB cost 1'], 0)
df1['AA-BB remain 2'] = np.where((df1['AA-BB remain 1'] - df1['Number of Seats']) >= 0, df1['AA-BB remain 1'] - df1['Number of Seats'], 0)
df1
```
